```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import pandas as pd

url = "https://raw.githubusercontent.com/AliAnusKhan/Assignement1_Flyrank/main/data/raw/content_refresh_anonymized.csv"
fly_rank = pd.read_csv(url)

print("Total rows:", len(fly_rank))
print(fly_rank.columns.tolist())
fly_rank.shape
fly_rank.head()

```

```python
import pandas as pd

# Key numbers for CTR / Engagement Opportunity Scoring lane

print("Total content pieces:", len(fly_rank))
print("Average CTR:", round(fly_rank['ctr'].mean(), 4))
print("Average position:", round(fly_rank['avg_position'].mean(), 2))

# Opportunity signal: high impressions but low CTR
high_impression_low_ctr = fly_rank[(fly_rank['impressions_90d'] > fly_rank['impressions_90d'].median()) &
                               (fly_rank['ctr'] < fly_rank['ctr'].median())]
print("Pages with high impressions but low CTR:", len(high_impression_low_ctr))
print("Percentage of total:", round(len(high_impression_low_ctr) / len(fly_rank) * 100, 1), "%")
```

The starter dataset contains 30,000 content pieces (one row = one piece of content per client). The average CTR across all content is 51.07%, with an average search ranking position of 16.34. Notably, 3,309 pages (11.0%) show above-median impressions but below-median CTR — indicating a meaningful, sizeable pool of content with untapped engagement opportunity. This confirms the lane is worth pursuing: there's enough signal and volume to build a scoring model that prioritizes these pages for optimization.
